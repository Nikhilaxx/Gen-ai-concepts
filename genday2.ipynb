```python
'''
perceptron->hidden layer
in forward propagation ,weights and bias values cant be changed and in backward propagation ,it can be changed

deep learning:-
can process large data and can perform multiple processes at the same time

Bias in Neural Networks:
Bias is a parameter in neural networks that helps the model make better predictions. 
It is an additional constant term added to the input of the activation function for each neuron, 
allowing the model to shift the activation curve left or right.
Without the bias, the model would have to pass through the origin (0, 0), limiting its ability to fit data accurately

*bias is an additional parameter added to the weighted sum of inputs to adjust the output,
 allowing the model to fit the data better. It helps shift the activation
 function and enables the network to make more flexible predictions, even when input values are zero.
'''
```

```python
'''

activation function ->in deep learning is a mathematical function applied to the output of a neuron (node)
in a neural network. It introduces non-linearity into the network, allowing the model to learn and solve
more complex tasks. Without an activation function, the neural network would behave like a simple linear 
regression model,
making it incapable of handling non-linear relationships

SLP: Simple, used for linearly separable problems, single layer.
MLP: More powerful, handles complex, non-linear problems, multiple layers with non-linear activation.
'''
```
