```python
import cv2
color_image=cv2.imread('rose.jpeg')
gray_image=cv2.cvtColor(color_image,cv2.COLOR_BGR2RGB)
cv2.imshow('COLOR IMAGE',color_image)
cv2.imshow('grayscale IMAGE',gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
image=cv2.imread('rose.jpeg')
blurred_image=cv2.GaussianBlur(image,(15,15),0)
cv2.imshow('original image',image)
cv2.imshow('blurred IMAGE',blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread("rose.jpeg", cv2.IMREAD_GRAYSCALE)
_,binary_segmentation = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt .figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(binary_segmentation, cmap='gray')
plt.title("Binary Segmentation")
plt.axis('off')
plt.show()
```

```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread("rose.jpeg", cv2.IMREAD_GRAYSCALE)
_, otsu = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt .figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(otsu, cmap='gray')
plt.title("Image Segmentation with otsu Thresholding")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np

# Load the two images
image1 = cv2.imread('1.jpg')
image2 = cv2.imread('2.jpg')

# Blend the images using addWeighted
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)

# Display the result
cv2.imshow('Weighted Image', weightedSum)

# Wait for a key press and close the window
if cv2.waitKey(0) & 0xFF == 27:  # 27 is the ASCII code for the 'Esc' key
    cv2.destroyAllWindows()
```
